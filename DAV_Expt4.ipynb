```python
import requests
import pandas as pd
import json #

```

```python
# Step 1: API endpoint for Open Food Facts
food_url = "https://world.openfoodfacts.org/cgi/search.pl"

# Parameters for searching for products containing "chocolate"
food_params = {
    "search_terms": "chocolate",
    "search_simple": 1,
    "action": "process",
    "json": 1,
    "page_size": 20 # Get 20 results
}

# Make the API request
response = requests.get(food_url, params=food_params)
data = response.json()

# Check if the request was successful and if there are products
if response.status_code == 200 and data.get('products'):
    # Load the products into a pandas DataFrame
    food_df = pd.DataFrame(data['products'])

else:
    print("Could not retrieve data from the Open Food Facts API.")
    print(f"Status Code: {response.status_code}")
    print(f"Response: {response.text}")
```

```python
# Save the DataFrame to a CSV file
food_df.to_csv("food_data.csv", index=False)

print("DataFrame successfully saved to food_data.csv")
```
