```python

corpus_string = corpus[0]
tokens = []
current_token = ""
for char in corpus_string:
    if char.isalnum():
        current_token += char
    elif current_token:
        tokens.append(current_token.lower())
        current_token = ""
if current_token:
    tokens.append(current_token.lower())


unigram_counts = {}
for token in tokens:
    unigram_counts[token] = unigram_counts.get(token, 0) + 1


print("\nUnigram counts:")
print(unigram_counts)
```

by bigram

```python
bigram_counts = {}
for i in range(1, len(tokens)):
    bigram = (tokens[i-1], tokens[i])
    bigram_counts[bigram] = bigram_counts.get(bigram, 0) + 1

print("\nBigram counts:")
print(bigram_counts)
```

```python
bigram_probabilities = {}
for bigram, count in bigram_counts.items():
    word1 = bigram[0]
    word2 = bigram[1]
    unigram_count_of_word1 = unigram_counts.get(word1, 0)
    if unigram_count_of_word1 > 0:
        bigram_probability = count / unigram_count_of_word1
        bigram_probabilities[bigram] = bigram_probability
    else:
        bigram_probabilities[bigram] = 0


print("\nBigram Probabilities:")
print(bigram_probabilities)
```

```python
target_sequence = ["eos", "you", "read", "a", "book", "eos"]
sequence_probability = 1.0

for i in range(1, len(target_sequence)):
    bigram = (target_sequence[i-1], target_sequence[i])
    bigram_probability = bigram_probabilities.get(bigram, 0)
    sequence_probability *= bigram_probability

print(f"\nProbability of the sequence '{' '.join(target_sequence)}': {sequence_probability}")
```

**Conclusion:**

This notebook demonstrates the process of calculating unigram, bigram, and trigram counts and probabilities from a small corpus. The bigram model provides probabilities for pairs of words, while the trigram model considers sequences of three words. Calculating the probability of a sequence of words using these models highlights how the context of preceding words impacts the likelihood of subsequent words, with higher-order models potentially capturing more complex dependencies but also suffering from sparsity with limited data, leading to zero probabilities for unseen sequences.
