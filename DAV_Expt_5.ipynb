```python

import pandas as pd
# Load dataset
df = pd.read_csv("emp.csv")
# Display first few rows
print(df.head())
# View the last 5 rows
df.tail()
# Get a summary of the DataFrame
df.info()
# Get basic statistical details for numeric columns
df.describe()
# List all column names
df.columns
# Get the shape of the DataFrame (rows, columns)
df.shape
```

```python
print(df.isnull().sum())
# Drop rows with any missing values
df.dropna()
```

```python
# Fill in missing salary values with the mean
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
# Fill missing Experience with median
df['Experience'].fillna(df['Experience'].median(), inplace=True)
# Fill missing Performance Score with the mode
df['Performance_Score'].fillna(df['Performance_Score'].mode()[0], inplace=True)


df[df.duplicated()]

df.drop_duplicates(inplace=True)

df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])
```

```python
display(df.isnull().sum())
display(df[df.duplicated()])
display(df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], kde=True)
plt.title('Distribution of Employee Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Plotting Age Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=True, bins=10, color='skyblue')
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Experience', y='Salary', hue='Department', palette='Set2')
plt.title('Experience vs. Performance Score')
plt.xlabel('Experience (Years)')
plt.ylabel('Performance Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Department', y='Salary', palette='viridis')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Joining_Date', y='Salary', hue='Department', marker='o', linewidth=2)
plt.title('Joining Date vs. Salary')
plt.xlabel('Joining Date')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Department', y='Salary', hue='Department', palette='viridis', legend=False)
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()
```

```python
# Plotting Department Distribution as a Pie Chart
department_counts = df['Department'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Employees by Department')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
