# Thresholding


```python
image.shape
```

```python
threshold_img = image.copy()
```

```python
image.shape[0]
```

```python
a = int(input("Enter the threshold value: "))
for i in range(threshold_img.shape[0]):
  for j in range(threshold_img.shape[1]):
    if image[i][j] < a:
      threshold_img[i][j]=0
    else:
      threshold_img[i][j]=255
```

```python
plt.imshow(threshold_img, cmap='gray')
plt.show()
```

Gray Level Slicing

```python
graylevel_slicing_img = image.copy()
```

```python
#grayleveslicing 1 with background
a = 20
b = 70
for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    for k in range(3):
      if image[i,j] <= b and image[i,j] >= a:
        graylevel_slicing_img[i,j]=255
      else:
        graylevel_slicing_img[i,j]= image[i,j]
```

```python
plt.imshow(graylevel_slicing_img ,cmap='gray')
plt.show()
```

```python
#grayleveslicing 2 with background
a = 20
b = 70
for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    for k in range(3):
      if image[i,j] <= b and image[i,j] >= a:
        graylevel_slicing_img[i,j]=image[i,j]
      else:
        graylevel_slicing_img[i,j]= 255
plt.imshow(graylevel_slicing_img ,cmap='gray')
plt.show()
```

```python
#grayleveslicing 3 without background
a = 20
b = 70
for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    for k in range(3):
      if image[i,j] <= b and image[i,j] >= a:
        graylevel_slicing_img[i,j]=255
      else:
        graylevel_slicing_img[i,j]= 0
plt.imshow(graylevel_slicing_img ,cmap='gray')
plt.show()
```

```python
#grayleveslicing 4 without background
a = 20
b = 70
for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    for k in range(3): # Iterate through R, G, B channels
      if image[i,j] <= b and image[i,j] >= a:
        graylevel_slicing_img[i,j]=0
      else:
        graylevel_slicing_img[i,j]= 255
plt.imshow(graylevel_slicing_img ,cmap='gray')
plt.show()
```

Contrast_Stretching


```python
contrast_stretching_img = image.copy()
```

```python
l=0.3
m =3
n=1
a=100
b=140
v =l*a
w = n*(b-a) + v
for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    if image[i,j] <= a:
      contrast_stretching_img[i,j]=l*image[i,j]
    elif image[i,j] >= b:
      contrast_stretching_img[i,j]=m*(image[i,j]-a) + v
    else:
      contrast_stretching_img[i,j]=n*(image[i,j]-b) + w
```
