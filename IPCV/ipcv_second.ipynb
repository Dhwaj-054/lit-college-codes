```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("image.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
cv2_imshow(img)
```

```python
#m,n=img.shape
```

```python
l=0.5
a=100
b=150
m=2
n=0.8
contrast_img=np.zeros_like(img, dtype=np.uint8)
v=l*a
w=v+m*(b-a)
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    if(0<img[i,j]<=a):
      contrast_img[i,j]=l*img[i,j]
    elif(a<img[i,j]<=b):
      contrast_img[i,j]=m*(img[i,j]-a) + v
    else:
      contrast_img[i,j]=n*(img[i,j]-b) + w
```

```python
cv2_imshow(contrast_img)
```

```python
print("Dynamic Range Compression")
```

```python
c = 255 / (np.log(1 + np.max(img)))
log_transformed_img = c * np.log(img + 1)

log_transformed_img = np.array(log_transformed_img, dtype=np.uint8)
```

```python
cv2_imshow(log_transformed_img)
```

```python
Dynamic_img=np.zeros(img.shape[:2],dtype="uint8")
c=20

for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    Dynamic_img[i,j]=c*np.log(1+abs(img[i,j]))

cv2_imshow(Dynamic_img)
cv2.waitKey(0)
```

```python
Dynamic_img=np.zeros(img.shape[:2],dtype="uint8")
c=41

for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    Dynamic_img[i,j]=c*np.log(1+abs(img[i,j]))

cv2_imshow(Dynamic_img)
cv2.waitKey(0)
```

```python
print("Power law transformation")
```

```python
gamma=float(input("Enter the value of gamma "))

power_img=np.zeros(img.shape[:2],dtype="uint8")

for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    power_img[i,j]=(img[i][j]**(gamma))


cv2_imshow(power_img)
cv2.waitKey(0)

```
