```python
# Step 3: TF-IDF on plot overview
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Fill missing values
df['overview'] = df['overview'].fillna('')

# Apply TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['overview'])

# Compute cosine similarity
cos_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Map movie titles to indices
indices = pd.Series(df.index, index=df['title_x'].str.lower().str.strip()).drop_duplicates()

# Recommendation function
def get_recommendations(title, cos_sim=cos_sim, k=10):
    idx = indices[title.lower().strip()]
    sim_scores = list(enumerate(cos_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:k+1]
    movie_indices = [i[0] for i in sim_scores]
    return df['title_x'].iloc[movie_indices]

# Example
get_recommendations("The Dark Knight Rises")
```

```python
# Step 3: TF-IDF on plot overview
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Fill missing values
df['overview'] = df['overview'].fillna('')

# Apply TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['overview'])

# Compute cosine similarity
cos_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Map movie titles to indices
indices = pd.Series(df.index, index=df['title_x'].str.lower().str.strip()).drop_duplicates()

# Recommendation function
def get_recommendations(title, cos_sim=cos_sim, k=10):
    idx = indices[title.lower().strip()]
    sim_scores = list(enumerate(cos_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:k+1]
    movie_indices = [i[0] for i in sim_scores]
    return df['title_x'].iloc[movie_indices]

# Example
get_recommendations("Inception")
```

```python
# Step 4: Metadata based recommender
from ast import literal_eval

# Apply literal_eval to convert string representations of lists/dictionaries
for col in ['cast', 'crew', 'keywords', 'genres']:
    df[col] = df[col].apply(lambda x: literal_eval(x) if isinstance(x, str) else [])


# Extract director name
def get_director(crew_list):
    if isinstance(crew_list, list):
        for person in crew_list:
            if isinstance(person, dict) and person.get('job') == 'Director':
                return person.get('name')
    return ''

# Get top n names (actors/keywords)
def top_n_names(obj_list, n=3):
    if isinstance(obj_list, list):
        return [i.get('name').replace(" ", "") for i in obj_list[:n] if isinstance(i, dict) and i.get('name')]
    return []

# Build metadata features
df['director'] = df['crew'].apply(get_director)
df['cast_names'] = df['cast'].apply(lambda x: top_n_names(x, 3))
df['keywords_names'] = df['keywords'].apply(lambda x: top_n_names(x, 5))
df['genres_names'] = df['genres'].apply(lambda x: top_n_names(x, len(x))) # Get all genres


# Create a "soup" feature
def create_soup(x):
    return ' '.join(x['keywords_names']) + ' ' + ' '.join(x['cast_names']) + ' ' + x['director'].replace(" ", "") + ' ' + ' '.join(x['genres_names'])

df['soup'] = df.apply(create_soup, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Apply Count Vectorizer
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df['soup'])

# Compute cosine similarity
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)

# Use same recommendation function but with metadata similarity
get_recommendations("Inception", cos_sim=cosine_sim2)
```

```python
# Task 1: Weighted director in soup (repeated 3 times)
def create_weighted_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + \
           (x['director'].replace(" ", "") + ' ') * 3 + \
           ' '.join([g['name'].replace(" ", "") for g in x['genres']])

df['weighted_soup'] = df.apply(create_weighted_soup, axis=1)
```

```python
count_weighted = CountVectorizer(stop_words='english')
count_matrix_weighted = count_weighted.fit_transform(df['weighted_soup'])

cosine_sim_weighted = cosine_similarity(count_matrix_weighted, count_matrix_weighted)


get_recommendations("Inception", cos_sim=cosine_sim_weighted)
```

```python
# Load SBERT model
sbert_model = SentenceTransformer('all-MiniLM-L6-v2')

# Fill missing overviews
df['overview'] = df['overview'].fillna('')

# Encode movie overviews
sbert_embeddings = sbert_model.encode(df['overview'], convert_to_tensor=False)
```

```python
# Compute cosine similarity matrix
cosine_sim_sbert = cosine_similarity(sbert_embeddings, sbert_embeddings)

# Example recommendation with SBERT
get_recommendations("Inception", cos_sim=cosine_sim_sbert)
```
