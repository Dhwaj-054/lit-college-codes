```python
#1.	Create a GUI form with the details like (Name, age(take a password field to hide age ), 
#city, gender (radiobutton), favourite game (check button) Etc) and print the details on the frame area.

import tkinter as tk
from tkinter import messagebox

# Function to display details on the frame
def submit_details():
    name = name_entry.get()
    age = age_entry.get()
    city = city_entry.get()
    gender = gender_var.get()
    
    # Collecting favorite games selected
    fav_games = []
    if football_var.get():
        fav_games.append("Football")
    if basketball_var.get():
        fav_games.append("Basketball")
    if cricket_var.get():
        fav_games.append("Cricket")
    
    # Display the details in a messagebox or on the frame
    result = f"Name: {name}\nAge: {age}\nCity: {city}\nGender: {gender}\nFavourite Games: {', '.join(fav_games)}"
    result_label.config(text=result)

# Create the main window
root = tk.Tk()
root.title("Details Form")

# Name label and entry
tk.Label(root, text="Name:").grid(row=0, column=0, padx=10, pady=10)
name_entry = tk.Entry(root)
name_entry.grid(row=0, column=1, padx=10, pady=10)

# Age label and password entry (to hide the input)
tk.Label(root, text="Age (hidden):").grid(row=1, column=0, padx=10, pady=10)
age_entry = tk.Entry(root, show='*')
age_entry.grid(row=1, column=1, padx=10, pady=10)

# City label and entry
tk.Label(root, text="City:").grid(row=2, column=0, padx=10, pady=10)
city_entry = tk.Entry(root)
city_entry.grid(row=2, column=1, padx=10, pady=10)

# Gender label and radio buttons
tk.Label(root, text="Gender:").grid(row=3, column=0, padx=10, pady=10)
gender_var = tk.StringVar(value="Male")  # Default value
tk.Radiobutton(root, text="Male", variable=gender_var, value="Male").grid(row=3, column=1, sticky=tk.W)
tk.Radiobutton(root, text="Female", variable=gender_var, value="Female").grid(row=3, column=2, sticky=tk.W)

# Favourite games label and check buttons
tk.Label(root, text="Favourite Game(s):").grid(row=4, column=0, padx=10, pady=10)
football_var = tk.BooleanVar()
basketball_var = tk.BooleanVar()
cricket_var = tk.BooleanVar()

tk.Checkbutton(root, text="Football", variable=football_var).grid(row=4, column=1, sticky=tk.W)
tk.Checkbutton(root, text="Basketball", variable=basketball_var).grid(row=4, column=2, sticky=tk.W)
tk.Checkbutton(root, text="Cricket", variable=cricket_var).grid(row=4, column=3, sticky=tk.W)

# Submit button
submit_button = tk.Button(root, text="Submit", command=submit_details)
submit_button.grid(row=5, column=1, pady=10)

# Label to display result
result_label = tk.Label(root, text="", fg="blue", justify="left")
result_label.grid(row=6, column=0, columnspan=4, padx=10, pady=10)

# Start the GUI event loop
root.mainloop()
```

```python
#2.	Create a GUI frame to find factorial of an input number. User is allowed to enter a number into the text field whose factorial 
#is to be determined. On pressing the button the value of the text 
#field is firstly converted into integer and then processed to find its factorial. The result will get displayed on 
#command prompt.

import tkinter as tk
from tkinter import messagebox

# Function to compute factorial
def factorial(num):
    if num == 0 or num == 1:
        return 1
    else:
        return num * factorial(num - 1)

# Function to handle button press
def compute_factorial():
    try:
        # Get the input from the text field
        num = int(entry.get())
        
        if num < 0:
            messagebox.showerror("Input Error", "Please enter a non-negative integer.")
        else:
            # Calculate the factorial
            result = factorial(num)
            print(f"Factorial of {num} is: {result}")
            result_label.config(text=f"Factorial of {num} is: {result}")
    except ValueError:
        messagebox.showerror("Input Error", "Please enter a valid integer.")

# Create the main window
root = tk.Tk()
root.title("Factorial Calculator")

# Create the input label and text field
tk.Label(root, text="Enter a number:").grid(row=0, column=0, padx=10, pady=10)
entry = tk.Entry(root)
entry.grid(row=0, column=1, padx=10, pady=10)

# Button to calculate factorial
calc_button = tk.Button(root, text="Calculate Factorial", command=compute_factorial)
calc_button.grid(row=1, column=0, columnspan=2, pady=10)

# Label to display the result
result_label = tk.Label(root, text="", fg="blue")
result_label.grid(row=2, column=0, columnspan=2, pady=10)

# Start the GUI event loop
root.mainloop()
```

```python
#3.	Create a Christmas tree using various shapes (Triangle, rectangle(for stem), circle(for decoration),  of in Canvas 
import tkinter as tk

# Function to draw the Christmas tree
def draw_christmas_tree(canvas):
    # Drawing the tree using a triangle (foliage)
    canvas.create_polygon(150, 50, 100, 150, 200, 150, fill="green", outline="black")
    canvas.create_polygon(150, 30, 80, 100, 220, 100, fill="green", outline="black")
    canvas.create_polygon(150, 10, 60, 60, 240, 60, fill="green", outline="black")
    
    # Drawing the tree trunk using a rectangle
    canvas.create_rectangle(140, 150, 160, 200, fill="brown", outline="black")
    
    # Drawing ornaments (circles) for decoration
    canvas.create_oval(130, 70, 150, 90, fill="red", outline="black")
    canvas.create_oval(170, 70, 190, 90, fill="yellow", outline="black")
    canvas.create_oval(120, 110, 140, 130, fill="blue", outline="black")
    canvas.create_oval(160, 110, 180, 130, fill="purple", outline="black")
    canvas.create_oval(140, 130, 160, 150, fill="white", outline="black")

# Create the main window
root = tk.Tk()
root.title("Christmas Tree")

# Create a Canvas widget
canvas = tk.Canvas(root, width=300, height=300)
canvas.pack()

# Call the function to draw the tree
draw_christmas_tree(canvas)

# Start the GUI event loop
root.mainloop()
```
