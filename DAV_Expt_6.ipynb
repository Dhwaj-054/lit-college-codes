```python

import pandas as pd
# Load dataset
df = pd.read_csv("emp.csv")
# Display first few rows
print(df.head())
# View the last 5 rows
df.tail()
# Get a summary of the DataFrame
df.info()
# Get basic statistical details for numeric columns
df.describe()
# List all column names
df.columns
# Get the shape of the DataFrame (rows, columns)
df.shape
```

```python
print(df.isnull().sum())
# Drop rows with any missing values
df.dropna()
```

```python
# Fill in missing salary values with the mean
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
# Fill missing Experience with median
df['Experience'].fillna(df['Experience'].median(), inplace=True)
# Fill missing Performance Score with the mode
df['Performance_Score'].fillna(df['Performance_Score'].mode()[0], inplace=True)


df[df.duplicated()]

df.drop_duplicates(inplace=True)

df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])
```

```python
display(df.isnull().sum())
display(df[df.duplicated()])
display(df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn

# Select only numeric columns from the dataframe
numeric_df = df.select_dtypes(include=['float64', 'int64'])
# Now, calculate the correlation matrix on the numeric columns
plt.figure(figsize=(10, 6))
sn.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
print("""Heatmap (Feature Correlation):

The heatmap shows the correlation between the numeric features in your dataset. The colors indicate the strength and direction of the correlation.

A value close to 1 (red) indicates a strong positive correlation (as one variable increases, the other tends to increase).
A value close to -1 (blue) indicates a strong negative correlation (as one variable increases, the other tends to decrease).
A value close to 0 (lighter colors) indicates a weak or no correlation.
From the heatmap, we can see that:

'Salary' and 'Experience' have a relatively strong positive correlation (around 0.86), which is expected as more experienced employees generally earn higher salaries.
'Age' and 'Experience' also show a strong positive correlation (around 0.94), which makes sense as older employees tend to have more experience.
'Performance_Score' shows a weak positive correlation with 'Salary' (around 0.44) and 'Experience' (around 0.45), suggesting that while performance might influence salary and experience, it's not as strong a relationship as age and experience or salary and experience.""")
```

```python
plt.figure(figsize=(8,5))
sn.violinplot(x="Department", y="Salary", data=df, palette="muted")
plt.title("Salary Distribution per Department")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()
```

```python
print("""Violin Plot (Salary Distribution per Department):

The violin plot shows the distribution of salaries for each department. The shape of the violin represents the density of salaries at different values. The wider sections indicate a higher concentration of employees at those salary levels.

From the violin plot, we can observe:

The Finance department appears to have a higher median salary and a wider range of salaries compared to other departments, indicating a potentially higher earning potential and more variation in pay within that department.
The IT, HR, and Marketing departments seem to have generally lower salary ranges and medians compared to Finance.
The shape of the violins gives you an idea of the distribution - for example, if a violin is wider at the top, it means more employees in that department earn higher salaries.""")
```
