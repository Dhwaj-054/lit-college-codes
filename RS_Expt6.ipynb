```python
def get_recommendations(products, max_price=None, category=None, min_rating=None, brand=None):
    """
    Filter products based on user-defined constraints

    Parameters:
    - products: list of product dictionaries
    - max_price: maximum price constraint (optional)
    - category: product category constraint (optional)
    - min_rating: minimum rating constraint (optional)
    - brand: brand name constraint (optional)

    Returns:
    - List of recommended products matching all constraints
    """
    recommendations = []

    for product in products:
        # Check each constraint
        price_match = max_price is None or product["price"] <= max_price
        category_match = category is None or product["category"].lower() == category.lower()
        rating_match = min_rating is None or product["rating"] >= min_rating
        brand_match = brand is None or product["brand"].lower() == brand.lower()

        # If all constraints are satisfied, add to recommendations
        if price_match and category_match and rating_match and brand_match:
            recommendations.append(product)

    return recommendations


def display_recommendations(recommendations):
    """Display recommended products in a formatted manner"""
    if recommendations:
        print(f"\n{'='*80}")
        print(f"✓ Found {len(recommendations)} Recommended Product(s):")
        print('='*80)

        for idx, product in enumerate(recommendations, 1):
            print(f"\n{idx}. {product['name']}")
            print(f"   Category: {product['category']}")
            print(f"   Brand: {product['brand']}")
            print(f"   Price: ₹{product['price']}")
            print(f"   Rating: {product['rating']} ⭐")
    else:
        print("\n" + "="*80)
        print("✗ No products found matching your constraints.")
        print("Try relaxing some constraints (higher price, lower rating, etc.)")
        print("="*80)

print("✓ Recommendation functions defined successfully")
```

```python
print("\n" + "="*80)
print("CONSTRAINT-BASED PRODUCT RECOMMENDATION SYSTEM")
print("="*80)

# Get user constraints
max_price = float(input("\nEnter maximum price (in ₹): "))
category = input("Enter product category (Electronics/Accessories/Home Appliances/Fashion): ")
min_rating = float(input("Enter minimum rating (0-5): "))

# Get recommendations
recommendations = get_recommendations(products, max_price, category, min_rating)

# Display results
display_recommendations(recommendations)
```

```python
import matplotlib.pyplot as plt

# Visualize product distribution by category
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Products by Category
category_counts = df_products['category'].value_counts()
axes[0, 0].bar(category_counts.index, category_counts.values, color='skyblue')
axes[0, 0].set_title('Product Distribution by Category', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('Category')
axes[0, 0].set_ylabel('Number of Products')
axes[0, 0].tick_params(axis='x', rotation=45)

# Plot 2: Price Distribution
axes[0, 1].hist(df_products['price'], bins=10, color='lightgreen', edgecolor='black')
axes[0, 1].set_title('Price Distribution', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('Price (₹)')
axes[0, 1].set_ylabel('Number of Products')

# Plot 3: Rating Distribution
axes[1, 0].hist(df_products['rating'], bins=10, color='coral', edgecolor='black')
axes[1, 0].set_title('Rating Distribution', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('Rating')
axes[1, 0].set_ylabel('Number of Products')

# Plot 4: Average Price by Category
avg_price_by_category = df_products.groupby('category')['price'].mean().sort_values()
axes[1, 1].barh(avg_price_by_category.index, avg_price_by_category.values, color='plum')
axes[1, 1].set_title('Average Price by Category', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('Average Price (₹)')
axes[1, 1].set_ylabel('Category')

plt.tight_layout()
plt.show()

print("\n✓ Visualizations generated successfully")
```

```python
# Example: Export recommendations to CSV
def export_recommendations(recommendations, filename="recommendations.csv"):
    """Export recommended products to CSV file"""
    if recommendations:
        df_recs = pd.DataFrame(recommendations)
        df_recs.to_csv(filename, index=False)
        print(f"\n✓ Recommendations exported to {filename}")
        print(f"Total records exported: {len(recommendations)}")
    else:
        print("\n✗ No recommendations to export")

# Example usage
test_recs = get_recommendations(products, max_price=30000, min_rating=4.5)
export_recommendations(test_recs, "high_rated_affordable_products.csv")
```
