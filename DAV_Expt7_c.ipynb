```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html, Input, Output
```

```python
f1_df = pd.read_csv('F1DriversDataset.csv')
display(f1_df.head())
f1_df.info()
```

```python
print(f1_df.isnull().sum())
```

```python
f1_df = f1_df.drop(columns=['Championship Years'])
print(f1_df.info())
```

```python
import ast

f1_df['Seasons'] = f1_df['Seasons'].apply(ast.literal_eval)
f1_df['Years_Active'] = f1_df['Seasons'].apply(lambda x: max(x) - min(x) + 1 if isinstance(x, list) and x else 0)
display(f1_df.head())
print(f1_df['Years_Active'].dtype)
```

```python
app = Dash(__name__)

app.layout = html.Div([
    html.H1("F1 Driver Dashboard", style={"textAlign": "center"}),

    html.Div([
        html.Label("Select Nationality:"),
        dcc.Dropdown(
            id="nationality_filter",
            options=[{"label": nationality, "value": nationality} for nationality in f1_df["Nationality"].unique()],
            value=None,
            placeholder="Select a nationality",
            multi=True
        )
    ], style={"width": "40%", "margin": "auto"}),

    dcc.Graph(id="races_vs_wins"),
    dcc.Graph(id="career_points_histogram"),
    dcc.Graph(id="years_active_by_nationality")
])
```

```python
@app.callback(
    [
        Output("races_vs_wins", "figure"),
        Output("career_points_histogram", "figure"),
        Output("years_active_by_nationality", "figure")
    ],
    [Input("nationality_filter", "value")]
)
def update_dashboard(selected_nationalities):
    filtered_df = f1_df.copy()
    if selected_nationalities:
        filtered_df = f1_df[f1_df["Nationality"].isin(selected_nationalities)]

    fig1 = px.scatter(filtered_df, x="Race_Entries", y="Race_Wins", color="Nationality", size="Podiums", hover_data=["Driver"], title="Race Entries vs Wins by Nationality")
    fig2 = px.histogram(filtered_df, x="Points", nbins=30, title="Distribution of Career Points", color="Nationality")
    fig3 = px.box(filtered_df, x="Nationality", y="Years_Active", color="Nationality", title="Years Active by Nationality")

    return fig1, fig2, fig3
```

```python
if __name__ == "__main__":
    app.run(debug=True, port=8050)
```
